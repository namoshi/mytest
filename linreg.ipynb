# 回帰分析

Pythonで利用するモジュールの読み込み

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

乱数の初期化

```python
np.random.seed(9876789)
```

データの生成

```python
nsample = 100
print(nsample)
#
x = np.linspace(0, 10, 100)
X = np.column_stack((x, x**2))
# model parameter
beta = np.array([1, 0.1, 10])
e = np.random.normal(size=nsample)
#
X = sm.add_constant(X)
y = np.dot(X, beta) + e
```

モデルの当てはめ（最小二乗法）と結果の表示

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

```python
print('X\n', X)
```

```python
print('y\n', y)
```

```python
print('Model Parameters: ', beta)
print('Estimated Parameters: ', results.params)
print('R2: ', results.rsquared)
```

```python
yest = X.dot(results.params)
```

```python
print('Estimated y\n', yest)
```

```python
err = y - yest
print('Errors\n', err)
```

```python
mse = sum(err ** 2)/len(err)
print('Mean Squared Errors =', mse)
```
